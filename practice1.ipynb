```python
import pandas as pd
melbourne_file_path = 'melb_data.csv'
melbourne_data = pd.read_csv(melbourne_file_path) 
melbourne_data
```

```python
melbourne_data.info
```

```python
melbourne_data.Address
```

```python
# two ways to access a value of a certain row and column, either done as below write in [] brackets first column 
# then row index

melbourne_data['Suburb'][13575]
```

```python
# or access it with at method
melbourne_data.at[13575,'Suburb']
```

```python
# if want to access any complete row of a dataframe use 'dataframe name.loc' for it and then in [] specify the 
# index value of the row you want to access 
melbourne_data.loc[13575]
```

```python
# cmd used to access a number of of rows in the given [x:y]
melbourne_data.loc[10000:10010]
```

```python
# cmds like dataframe name.head(x) or dataframe name.tail(x) can either show the first x rows in a dataframe 
# (.head(x)) or show the last x rows of dataframe (.tail(x))
melbourne_data.head(10)
```

```python
# the cmd below gives the index of the first not NAN value in a column. NAN value is basically a not given 
# value in a dataframe that is automatically then represented as a NAN in PANDAS.
melbourne_data.BuildingArea.first_valid_index()
```

```python
# cmd used to access a given ranged random sample of the dataframe
melbourne_data.sample(6)
```

```python
# cmd used to make a copy of the given dataframe
melbourne_data.copy()
```
